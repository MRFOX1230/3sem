## Задание 1
Для каждого артиста вывести все жанры, которые есть в его песнях, и для каждого жанра вывести наиболее продолжительную песню этого артиста.

```python
%%sql
SELECT ArtistName, TrackName, GenreName
FROM
(SELECT ArtistName, TrackName, GenreName, MAX(Milliseconds) as "TrackTime"
FROM
(
  SELECT artists.Name as ArtistName, tracks.Name as TrackName, genres.Name as GenreName,  Milliseconds
  FROM artists
  INNER JOIN albums, tracks, genres
  ON artists.ArtistID = albums.ArtistID and albums.AlbumID = tracks.AlbumID and tracks.GenreID = genres.GenreID)
GROUP BY ArtistName, GenreName)
```

## Задание 2
Посчитать скользящее среднее выручки по годам.


```python
%%sql
SELECT Year, Expense, AVG(Expense) OVER (ORDER BY Year) as RollAVG
FROM
(
  SELECT strftime('%Y', InvoiceDate) as Year, SUM(UnitPrice*Quantity) as Expense
  FROM
  (
    SELECT *
    FROM invoice_items
    INNER JOIN invoices
    ON invoice_items.InvoiceID = invoices.InvoiceID
  )
  GROUP BY Year
)
```

## Задание 3

Для каждого плейлиста вывести трек с наибольшей продолжительностью, трек с наименьшей стоимостью (если у треков одинаковая стоимость, то вывести тот, который весит меньше всех), и жанр данного плейлиста (жанром плейлиста будет называться тот жанр, треков которого больше всех в плейлисте).

```python
%%sql
SELECT Name as Playlist,
  Time as TheLongerTrack,
  TheCheaperTrack,
  FIRST_VALUE(Genre) OVER (PARTITION BY Name ORDER BY Count DESC) Genre
FROM (
  SELECT DISTINCT playlists.Name,
                  FIRST_VALUE(tracks.Name) OVER (PARTITION BY PlaylistId ORDER BY Milliseconds DESC) Time,
                  FIRST_VALUE(tracks.Name) OVER (PARTITION BY PlaylistId ORDER BY UnitPrice, Bytes) TheCheaperTrack,
                  genres.Name as Genre,
                  COUNT(*) OVER (PARTITION BY PlaylistId, GenreId) Count
  FROM tracks
  INNER JOIN playlist_track USING(TrackId)
  INNER JOIN playlists USING(PlaylistId)
  INNER JOIN genres USING(GenreId)
)
GROUP BY Name;
```

## Задание 4

Для каждого жанра вывести треки, которые встречались в большем числе заказов,
сумму, потраченную на данные треки, количество купленных треков и компанию, которая больше всех приобрела данную композицию.

```python
%%sql
SELECT Genre, Name as Track, TotalSum, TracksCount, Company
FROM (
  SELECT *
  FROM (
    SELECT DISTINCT TrackId,
                    tracks.Name,
                    Company,
                    genres.Name as Genre,
                    SUM(Total) OVER (PARTITION BY TrackId) TotalSum,
                    SUM(Quantity) OVER (PARTITION BY TrackId) TracksCount,
                    SUM(Quantity) OVER (PARTITION BY TrackId, Company) CompanyCount
    FROM tracks
    INNER JOIN invoice_items USING(TrackId)
    INNER JOIN genres USING(GenreId)
    INNER JOIN invoices USING(InvoiceId)
    INNER JOIN customers USING(CustomerId)
    ORDER BY TracksCount DESC,
          CASE
            WHEN Company is NULL THEN 1
            ELSE 0
          END,
          CompanyCount DESC
  )
  GROUP BY TrackId
  ORDER BY TracksCount DESC,
           CASE
              WHEN Company is NULL THEN 1
              ELSE 0
           END,
           CompanyCount DESC
)
GROUP BY Genre
```

## Задание 5
Составить рейтинг клиентов по количеству потраченных ими денег.

```python
%%sql
SELECT customers.FirstName || ' ' || customers.LastName AS CustomerName,
       SUM(invoice_items.UnitPrice * invoice_items.Quantity) AS TotalSpent
FROM customers
INNER JOIN invoices USING(CustomerId)
INNER JOIN invoice_items USING(InvoiceId)
GROUP BY customers.CustomerId
ORDER BY TotalSpent DESC
```
