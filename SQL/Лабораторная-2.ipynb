## Задание 1
Вывести названия всех групп, их песен и названия их альбомов для всех треков жанра Rock, приобретенные сотрудниками Microsoft.


```python
%%sql
SELECT artists.Name as "Группа", tracks.Name as "Песня", albums.Title as "Альбом"
FROM genres

INNER JOIN tracks, albums, invoice_items, invoices, customers, artists
ON genres.GenreID = tracks.GenreID and tracks.TrackID = invoice_items.TrackID and
invoice_items.InvoiceID = invoices.InvoiceID and invoices.CustomerID = customers.CustomerID and
tracks.AlbumID = albums.AlbumID and albums.ArtistId = artists.ArtistID

WHERE genres.Name = "Rock" and Company = "Microsoft Corporation"
```

## Задание 2
Вывести 10 самых больших по размеру треков жанра Blues и формата MPEG (используйте команду LIMIT 10). Только название и автор.


```python
%%sql
SELECT tracks.Name as "Песня", tracks.Composer as "Автор", Bytes as "Объем файла"
FROM genres

INNER JOIN tracks, media_types
ON genres.GenreID = tracks.GenreID and tracks.MediaTypeID = media_types.MediaTypeID
WHERE genres.Name = "Blues" and media_types.Name = "MPEG audio file"
ORDER BY Bytes DESC
LIMIT 10
```

## Задание 3
Для каждой компании вывести общее количество купленных джазовых композиций.


```python
%%sql
SELECT Company as "Компания", COUNT(Company) as "Кол-во купленных джазовых композиций"
FROM genres

INNER JOIN tracks, invoice_items, invoices, customers
ON genres.GenreID = tracks.GenreID and tracks.TrackID = invoice_items.TrackID and
invoice_items.InvoiceID = invoices.InvoiceID and invoices.CustomerID = customers.CustomerID
WHERE genres.Name = "Jazz" and Company is not NULL
GROUP BY Company
```

## Задание 4
Вывести топ 3 альбома по продолжительности (продолжительность альбома - сумма длин треков в нём). Только название и продолжительность.


```python
%%sql
SELECT Title as "Альбом", SUM(Milliseconds) as "Продолжительность в мс"
FROM albums

INNER JOIN tracks
ON albums.AlbumID = tracks.AlbumID
GROUP BY Title
ORDER BY SUM(Milliseconds) DESC
LIMIT 3
```

## Задание 5
Для каждого набора (жанр, тип медиа) вывести среднюю цену по стоимости трека и их общее количество, причем вывести только те наборы, для которых **все** треки стоят больше 1,5$ и для которых в наборе есть хотя бы один трек.

```python
%%sql
SELECT media_types.Name as "Формат файла", genres.Name as "Жанр",
AVG(UnitPrice) as "Средняя цена треков", COUNT(genres.Name) as "Кол-во треков"
FROM genres

INNER JOIN tracks, media_types
ON genres.GenreID = tracks.GenreID and
media_types.MediaTypeID = tracks.MediaTypeID

GROUP BY genres.Name, media_types.Name
HAVING COUNT(genres.Name) >= 1 and MIN(UnitPrice) > 1.5
```
